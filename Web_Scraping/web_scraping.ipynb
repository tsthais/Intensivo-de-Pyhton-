# Automação Web e Busca de Informações com Python

```python
# Bibliotecas utilizadas

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Abrir um navegador

pasta = "C:\\Users\\thais\\Documents\\chrome\\chromedriver"
navegador = webdriver.Chrome(pasta)
```

```python
# Passo 1 

# Pegar cotação do dólar

# Abrir o site do google

navegador.get("https://www.google.com/")

# Procurar barra de pesquisa no site do google

barra_pesquisa = navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input')

# Enviar pesquisa desejada

barra_pesquisa.send_keys("cotação dólar")

# Apertar enter para pesquisar

pesquisar = navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input')

pesquisar.send_keys(Keys.ENTER)

# Coletando o preço do dólar

dolar = navegador.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')

cotacao_dolar = dolar.get_attribute("data-value")
```

```python
# Passo 2

# Cotação do euro

# Abrir o site do google

navegador.get("https://www.google.com/")

# Procurar barra de pesquisa no site do google

barra_pesquisa = navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input')

# Enviar pesquisa desejada

barra_pesquisa.send_keys("cotação euro")

# Apertar enter para pesquisar

pesquisar = navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input')

pesquisar.send_keys(Keys.ENTER)

# Coletando o preço do euro

euro = navegador.find_element(By.XPATH, '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')

cotacao_euro = euro.get_attribute("data-value")
```

```python
# Passo 3

# Cotação do ouro

# Abrir primeiro site na busca de cotação do ouro

navegador.get("https://www.melhorcambio.com/ouro-hoje")

# Pegar o valor do ouro que aparece no site 

ouro = navegador.find_element(By.XPATH, '//*[@id="comercial"]')

cotacao_ouro = ouro.get_attribute("value")

# Tranformando vírgula para ponto

cotacao_ouro = cotacao_ouro.replace(",", ".")

# Sair do navegador 

navegador.quit()
```

```python
# Passo 4

# Importar base de dados 

tabela = pd.read_excel("Produtos.xlsx")

# Apresentando tabela 

display(tabela)

# Atualizar os dados 

tabela.loc[tabela['Moeda'] == "Dólar", "Cotação"] = float(cotacao_dolar)
tabela.loc[tabela['Moeda'] == "Euro", "Cotação"] = float(cotacao_euro)
tabela.loc[tabela['Moeda'] == "Ouro", "Cotação"] = float(cotacao_ouro)

# Apresentar diferenças

display(tabela)
```

```python
# Passo 5

# Calcular novos valores

tabela['Preço de Compra'] = tabela['Preço Original']*tabela['Cotação']

tabela['Preço de Venda'] = tabela['Preço de Venda']*tabela['Margem']

display(tabela)

# Exportar os novos dados

tabela.to_excel("Produtos Novo.xlsx", index=False)
```
